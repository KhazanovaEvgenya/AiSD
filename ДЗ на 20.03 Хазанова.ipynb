Класс – шар. Методы – расчет площади и объема шара. Поля – координаты центра (x, y, z), радиус шара.

```python
class Ball:
    from math import pi

    def __init__(self, x, y, z, radius):
        self.x = x
        self.y = y
        self.z = z
        self.radius = radius

    def area(self):    # площадь шара
        return 4 * self.pi * self.radius ** 2

    def volume(self):    #объем шара
        return (4/3) * self.pi * self.radius ** 3


# Создаем экземпляры класса Ball
ball1 = Ball(x=0, y=0, z=0, radius=1)  # Шар с центром в (0,0,0) и радиусом 1
ball2 = Ball(x=1, y=2, z=3, radius=4)  # Шар с центром в (1,2,3) и радиусом 4

# Тестируем метод расчета площади поверхности
print("Площадь поверхности ball1:", ball1.area())
print("Площадь поверхности ball2:", ball2.area())

# Тестируем метод расчета объема
print("Объем ball1:", ball1.volume())
print("Объем ball2:", ball2.volume())
```

    Составьте описание класса многочлена вида ax+b. Включите в класс методы: вывода описания многочлена на экран, вычисления значения многочлена для заданного аргумента, вычисления корня линейного уравнения с проверкой неравенства коэффициента b нулю, а также свойство, позволяющее возвести многочлен в квадрат.

```python
class Line:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def __str__(self):
        return f"{self.a}x + {self.b}"

    def calculate(self, x):
        """
        Вычисляет значение многочлена для заданного аргумента x.
        """
        return self.a * x + self.b

    def root(self):
        """
        Вычисляет корень многочлена, если это возможно.
        """
        if self.a == 0:
            return "Корень не может быть найден, a = 0."
        return -self.b / self.a

    @property
    def squared(self):
        # Возведение в квадрат даёт многочлен вида (a^2)x^2 + (2ab)x + b^2
        return Line(self.a ** 2, 2 * self.a * self.b + self.b ** 2)



one = Line(a=2, b=3)  # Многочлен 2x + 3
two = Line(a=0, b=5)  # Многочлен 0x + 5

# Выводим многочлены на экран
print(f"Многочлен 1: {one}")
print(f"Многочлен 2: {two}")

# Вычисляем значение многочлена при x = 10
x = 10
print(f"Значение one при x = {x}: {one.calculate(x)}")
print(f"Значение two при x = {x}: {two.calculate(x)}")

# Находим корень многочлена
print(f"Корень многочлена one: {one.root()}")
print(f"Корень многочлена two: {two.root()}")  

# Возводим многочлен в квадрат и выводим результат
print(f"Многочлен poly1, возведенный в квадрат: {one.squared}")
```

Создайте класс ПРОГРАММНОЕ ОБЕСПЕЧЕНИЕ с методами, позво-ляющими вывести на экран информацию о программном обеспечении, а также определить соответствие возможности использования (на текущую дату). Создайте дочерние классы СВОБОДНОЕ (название, производитель), УСЛОВНО БЕСПЛАТНОЕ (название, производитель, дата установки, срок бесплатного использования), КОММЕРЧЕСКОЕ (название, производитель, цена, дата установки, срок использования) со своими методами вывода информации на экран и определения возможности использования на текущую дату. Создайте список из видов программного обеспечения, выведите полную информацию из базы на экран, а также организуйте поиск программного обеспечения, которое допустимо использовать на текущую дату

```python
class Software:

    def __init__(self, name, manufacturer):
        """
        Инициализирует новый экземпляр программного обеспечения.
        """
        self.name = name
        self.manufacturer = manufacturer

    def display_info(self):
        """
        Выводит информацию о программном обеспечении.
        """
        return f"Название: {self.name}, Производитель: {self.manufacturer}"

    def is_usable(self):
        """
        Определяет, доступно ли программное обеспечение для использования.
        """
        return True

class FreeSoftware(Software):
    """
    Класс для представления свободного программного обеспечения.
    """

    def display_info(self):
        """
        Выводит информацию о свободном программном обеспечении.
        """
        info = super().display_info()
        return f"{info} (Свободное ПО)"


from datetime import datetime, timedelta

class FreemiumSoftware(Software):
    """
    Класс для представления условно-бесплатного программного обеспечения.
    """

    def __init__(self, name, manufacturer, install_date, free_period):
        """
        Инициализирует новый экземпляр условно-бесплатного программного обеспечения.
        """
        super().__init__(name, manufacturer)
        self.install_date = datetime.strptime(install_date, "%Y-%m-%d")
        self.free_period = timedelta(days=free_period)

    def display_info(self):
        """
        Выводит информацию об условно-бесплатном программном обеспечении.
        """
        info = super().display_info()
        expiry_date = self.install_date + self.free_period
        return f"{info}, Дата установки: {self.install_date.date()}, Срок бесплатного использования до: {expiry_date.date()}"

    def is_usable(self):
        """
        Определяет, доступно ли условно-бесплатное программное обеспечение для использования.
        """
        return datetime.now() < self.install_date + self.free_period


class CommercialSoftware(Software):
    """
    Класс для представления коммерческого программного обеспечения.
    """

    def __init__(self, name, manufacturer, price, install_date, usage_period):
        """
        Инициализирует новый экземпляр коммерческого программного обеспечения.
        """
        super().__init__(name, manufacturer)
        self.price = price
        self.install_date = datetime.strptime(install_date, "%Y-%m-%d")
        self.usage_period = timedelta(days=usage_period)

    def display_info(self):
        """
        Выводит информацию о коммерческом программном обеспечении.
        """
        info = super().display_info()
        expiry_date = self.install_date + self.usage_period
        return f"{info}, Цена: {self.price}, Дата установки: {self.install_date.date()}, Срок использования до: {expiry_date.date()}"

    def is_usable(self):
        """
        Определяет, доступно ли коммерческое программное обеспечение для использования.
        """
        return datetime.now() < self.install_date + self.usage_period


software_list = [
    FreeSoftware("VS Code", "Microsoft"),
    FreemiumSoftware("Sublime Text", "Sublime HQ", "2023-01-01", 30),
    CommercialSoftware("Photoshop", "Adobe", 999, "2023-01-01", 365),
]

print("Вся доступная информация о ПО:")
for software in software_list:
    print(software.display_info())

print("\nДоступное для использования ПО на сегодня:")
for software in software_list:
    if software.is_usable():
        print(software.display_info())
```
